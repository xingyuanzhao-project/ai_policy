```python
import pandas as pd
df = pd.read_csv('cleaned_data.csv', dtype={'chr': str})
```

```python
# Reshape dummy variables into one categorical column
stage_cols = ['Introduced', 'In committee', 'In cross chamber', 'In cross committee', 'Passed', 'Signed']

# Find which stage column has value 1 for each row
df['Legislative Process Detailed'] = df[stage_cols].apply(
    lambda row: next((col for col in stage_cols if row[col] == 1), None), axis=1
)

# Reorder columns to insert the new column after 'Legislative process'
cols = df.columns.tolist()
leg_process_idx = cols.index('Legislative process')
new_cols = cols[:leg_process_idx+1] + ['Legislative Process Detailed'] + cols[leg_process_idx+1:-1]
df_reshaped = df[new_cols]
```

```python
df_by_legislative_process = df.groupby(['Jurisdiction', 'Legislative process']).size().reset_index(name='Count')
df_by_legislative_process.to_csv('by_legislative_process.csv', index=False)
```

```python
# Filter out inactive bills and combine active bills + laws signed
df_active = df[df['Legislative process'] != 'INACTIVE BILLS']
df_active_by_state = df_active.groupby('Jurisdiction').size().reset_index(name='Count')
df_active_by_state.to_csv('active_by_state.csv', index=False)
```
